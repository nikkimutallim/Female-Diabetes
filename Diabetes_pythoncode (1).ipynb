```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/mutal/OneDrive - University of Iowa/Desktop/Statitistics and Data Visualisation/diabetes.csv")
```

```python
fig, ax = plt.subplots(figsize = (25, 25))
sns.set(font_scale=2.5)
sns.heatmap(ax=ax,
            data=data.corr(), 
            annot=True, 
            cmap='Blues', 
            annot_kws={'size': 20}, 
            robust=True).set(title="Correlation of Variables")
plt.savefig('saving-a-high-resolution-seaborn-plot.png')
```
